```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('imdb_movies.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['date_x'].tail()
```

```python
fig = px.scatter(df, x='budget_x', y='revenue', title='Budget vs Revenue')
fig.show()
```

```python
fig = px.pie(df.sort_index(ascending=True).head(190), names='orig_lang', title='Original Language')
fig.show()
```

```python
fig = px.scatter(df, x='orig_lang', y='revenue', title='Original Language vs Revenue')
fig.show()
```

```python
fig = px.pie(df.sort_index(ascending=True).head(50), names='genre', title='Genre')
fig.show()
```

```python
# find the top 10 movies with the highest revenue
fig = px.bar(df.sort_index(ascending=False).head(10), x="date_x", y="revenue", title="Top 10 Movies Revenue Over Time", hover_name="orig_title")
fig.show()
```

```python
# find the top 10 movies with the highest budget
fig = px.bar(df.sort_index(ascending=False).head(10), x="date_x", y="budget_x", title="Top 10 Movies Budget Over Time", hover_name="orig_title")
fig.show()
```

```python
# find genre with the highest revenue
fig = px.bar(df.groupby('genre')['revenue'].sum().sort_values(ascending=False).head(10), title='Genre with the Highest Revenue')
fig.show()
```

```python
# find genre with the highest budget
fig = px.bar(df.groupby('genre')['budget_x'].sum().sort_values(ascending=False).head(10), title='Genre with the Highest Budget')
fig.show()
```

```python
# find original language with overall highest revenue
fig = px.bar(df.groupby('orig_lang')['revenue'].sum().sort_values(ascending=False).head(10), title='Original Language with the Highest Revenue')
fig.show()
```

```python
# find original language with overall highest budget
fig = px.bar(df.groupby('orig_lang')['budget_x'].sum().sort_values(ascending=False).head(10), title='Original Language with the Highest Budget')
fig.show()
```

```python
# find original title with the highest revenue
fig = px.bar(df.sort_values(by='revenue', ascending=False).head(10), x='orig_title', y='revenue', title='Original Title with the Highest Revenue')
fig.show()
```

```python
# find original title with the highest budget
fig = px.bar(df.sort_values(by='budget_x', ascending=False).head(10), x='orig_title', y='budget_x', title='Original Title with the Highest Budget')
fig.show()
```

```python
# find country with creative movies
fig = px.bar(df.groupby('country')['orig_title'].count().sort_values(ascending=False).head(10), title='Country with Creative Movies')
fig.show()
```

```python
# find country with most revenue
fig = px.bar(df.groupby('country')['revenue'].sum().sort_values(ascending=False).head(10), title='Country with Most Revenue')
fig.show()
```

```python
# genre with most movies
fig = px.bar(df.groupby('genre')['orig_title'].count().sort_values(ascending=False).head(10), title='Genre with Most Movies')
fig.show()
```

```python
# names with similar movies
fig = px.bar(df.groupby('names')['genre'].count().sort_values(ascending=False).head(10), title='Names with Similar Movies')
fig.show()
```

```python
# top 10 movies with their similar overview names
fig = px.scatter(df.sort_values(by='revenue', ascending=False).head(), x='overview', y='revenue', title='Top 5 Movies with Similar Overview Names')
fig.show()
```

```python
# movie recommendation based on revenue
fig = px.scatter(df.sort_values(by='revenue', ascending=False).head(10), x='revenue', y='orig_title', title='Movie Recommendation Based on Revenue')
fig.show()
```

```python
# movie recommendation based on genre
fig = px.scatter(df.sort_values(by='genre', ascending=False).head(10), x='genre', y='orig_title', title='Movie Recommendation Based on Top 10 Genre')
fig.show()
```
